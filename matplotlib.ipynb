# グラフの基本的な書き方

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y)
plt.show()
```

```python
#x,yの要素の数が違うとエラー
#x = [1,2,3,4,5]
#y = [1,4,9,16]
#plt.plot(x,y)
#plt.show()
```

```python
#x,y軸に名前（ラベル）の表示
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#x,y軸に名前（ラベル）の大きさの変更
#fontsizeに数値で指定
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y)
plt.xlabel('X', fontsize=18)
plt.ylabel('Y', fontsize=18)
plt.show()
```

```python
#ラベルの背景にグリッド(grid)を表示
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y)
plt.xlabel('X', fontsize=18)
plt.ylabel('Y', fontsize=18)
plt.grid()
plt.show()
```

```python
#線グラフの色を変更
#plot()の中にcolorで色を指定
x = [1,2,3,4,5]
y = [1,4,9,16,25]
#colorの色指定は色の頭文字をとってもできる（redなら'r',blueなら'b',blackなら'k'）
#色はたくさんあるので調べてみると良い（カラーコードでもできる）
plt.plot(x,y, color='red')
plt.xlabel('X', fontsize=18)
plt.ylabel('Y', fontsize=18)
plt.grid()
plt.show()
```

```python
#線の太さを変更する
#plot()に linewidth=数値 で指定
#指定した数値が大きいほど線が太くなる
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y, linewidth=5)
plt.xlabel('X', fontsize=18)
plt.ylabel('Y', fontsize=18)
plt.grid()
plt.show()
```

```python
#線のスタイルを変更する
#plot()にlinestyle=''で指定（デフォルトはsolid）
#dashed（破線）,dashdot（破線と点線が混じった線）で指定できる
#':'でも'--'でも指定可能
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y, linestyle='--')
plt.xlabel('X', fontsize=18)
plt.ylabel('Y', fontsize=18)
plt.grid()
plt.show()
```

```python
#xとyが交わる点にマーカーをつける
#plot()にmarker=''で指定
x = [1,2,3,4,5]
y = [1,4,9,16,25]
#'o'は小文字のo 0ではない
plt.plot(x,y, marker='o')
plt.xlabel('X', fontsize=18)
plt.ylabel('Y', fontsize=18)
plt.grid()
plt.show()
```

```python
#グラフに注釈を加える
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y, marker='o')
plt.xlabel('X', fontsize=18)
plt.ylabel('Y', fontsize=18)
plt.grid()
#注釈のテキスト、注釈を表示する位置、注釈テキストの一文字目の位置、注釈テキストのサイズ、注釈テキストのカラー、矢印の色を指定
plt.annotate('annotation', xy=(3,9), xytext=(1,20), fontsize=15, color='r', arrowprops=dict(color='gray'))
plt.show()
```

```python
#グラフのタイトルの表示
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y, marker='o')
plt.xlabel('X', fontsize=18)
plt.ylabel('Y', fontsize=18)
plt.grid()
#注釈のテキスト、注釈を表示する位置、注釈テキストの一文字目の位置、注釈テキストのサイズ、注釈テキストのカラー、矢印の色を指定
plt.annotate('annotation', xy=(3,9), xytext=(1,20), fontsize=15, color='r', arrowprops=dict(color='gray'))
#グラフのタイトルは日本語フォントをインストールすれば表示可能
plt.title('graph title')
plt.show()
```

# グラフの重ね書き

```python
#numpyは高水準の数学関数。多次元配列も高速に計算可能
import numpy as np
```

```python
#array配列にすることで、配列に対して一度に四則演算できる
arr_x = np.array([1,2,3,4,5])
arr_x/2
```

```python
#listではnumpyと同じように計算できない
#list_x = [1,2,3,4,5]
#list_x/2
```

```python
#グラフを3つ重ねて表示
#描きたいグラフのplt.plot()を重ねる
x = [1,2,3,4,5]
y = np.array([1,4,9,16,25])
plt.plot(x,y, color='r')
plt.plot(x,y/2, color='b')
plt.plot(x,y/3, color='g')
plt.xlabel('X', fontsize=18)
plt.ylabel('Y', fontsize=18)
plt.grid()
plt.title('graph title')
plt.show()
```

```python
#グラフの凡例の表示
#plot()にlabel=''でラベル名を指定
#plt.legend()を記述
x = [1,2,3,4,5]
y = np.array([1,4,9,16,25])
plt.plot(x,y, color='r', label='A')
plt.plot(x,y/2, color='b', label='B')
plt.plot(x,y/3, color='g', label='C')
plt.xlabel('X', fontsize=18)
plt.ylabel('Y', fontsize=18)
plt.grid()
plt.title('graph title')
plt.legend()
plt.show()
```

```python
#凡例のフォントサイズの指定
#legend()にfontsize='数値'で指定
x = [1,2,3,4,5]
y = np.array([1,4,9,16,25])
plt.plot(x,y, color='r', label='A')
plt.plot(x,y/2, color='b', label='B')
plt.plot(x,y/3, color='g', label='C')
plt.xlabel('X', fontsize=18)
plt.ylabel('Y', fontsize=18)
plt.grid()
plt.title('graph title')
plt.legend(fontsize=15)
plt.show()
```

```python
#作成したグラフの保存
#plt.savefig('保村したいファイル名.png or .jpg or .pdf')
x = [1,2,3,4,5]
y = np.array([1,4,9,16,25])
plt.plot(x,y, color='r', label='A')
plt.plot(x,y/2, color='b', label='B')
plt.plot(x,y/3, color='g', label='C')
plt.xlabel('X', fontsize=18)
plt.ylabel('Y', fontsize=18)
plt.grid()
plt.title('graph title')
plt.legend(fontsize=15)
plt.savefig('graph.png')
plt.show()
```
